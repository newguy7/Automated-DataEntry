```python
data = pd.read_csv('amazon_reviews.csv')
data.shape
```

```python
# checking nan values
data.isna().sum()
```

```python
data = data.dropna()
data.isna().sum()
```

```python
Y = data['star_rating']
X = data[['product_title', 'product_category','review_headline', 'review_body']]
print(X.shape)
print(Y.shape)
```

```python
import numpy as np
import re
import nltk
import string

# checking nan values
X.isna().sum()
```

```python
# Text Preprocessing

#Lower Casing
X = X.applymap(str.lower)
X.head()
```

```python
# Removal of URLs
# before removing punctuations
def remove_urls(text):
  url_pattern = re.compile(r'https?://\S+')
  return url_pattern.sub(r'',text)

for k in X.columns:
  X[k]= X[k].apply(lambda text: remove_urls(text))

X.head()
```

```python
# Removal of Punctuations
PUNCT_TO_REMOVE = string.punctuation

# The first two arguments '' and '' represent the characters we want to replace and their corresponding replacements.
# if both are empty it indicates that we want to remove characters rather than replacing them
# str.maketrans('', '', PUNCT_TO_REMOVE)
# creates a translation table where each character in PUNCT_TO_REMOVE is mapped to None,
# effectively removing those characters from the text

def remove_punctuation(text):
  return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

# Applying to each columns
for k in X.columns:
  X[k]= X[k].apply(lambda text: remove_punctuation(text))
X.head()
```

```python
# Removal of stopwords ( 'the', 'a' etc)
import nltk
from nltk.corpus import stopwords

# Downloading package stopwords
nltk.download('stopwords')

# set will remove any duplicate words
STOPWORDS = set(stopwords.words('english'))


def remove_stopwords(text):
  return " ".join([word for word in str(text).split() if word not in STOPWORDS])

for k in X.columns:
  X[k]= X[k].apply(lambda text: remove_stopwords(text))

X.head()
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

for k in X.columns:
  X[k]=X[k].apply(lambda text: lemmatize_words(text))
```
