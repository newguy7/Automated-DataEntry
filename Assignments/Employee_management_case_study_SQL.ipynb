```python
'''
Create a Employee Management Case Study
Which Should have
1) Add Employee
2) Remove Employee by id
3) Remove employee by name
4) Update employee by id
5) Update employee by name
6) display all
7) Display By ID
8) Exit
'''
```

```python
import mysql.connector

class Employee:

  def __init__(self,name:str,age:int,id:int,position:str,salary:float):
    self.name = name
    self.age = age
    self.id = id
    self.position = position
    self.salary = salary

  def display_info(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Id: {self.id}")
    print(f"Position: {self.position}")
    print(f"Salary: {self.salary}")
    print("******************")

class EmployeeManagement:
    def __init__(self):
        self.connection = self.start_connection()
        self.create_employee_table()

    def start_connection(self):
        # creating connection with MySQL
        try:
            connection = mysql.connector.connect(
                host="localhost",
                user="root",
                passwd="hello",
                database = "Employee"
            )
            print("Successfully connected to MySQL")

            # Create the database if it doesn't exist
            cursor = connection.cursor()
            cursor.execute("CREATE DATABASE IF NOT EXISTS Employee")
            cursor.close()

            return connection
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return None

    def create_employee_table(self):
        # creating connection with MySQL
        try:
            cursorObject = self.connection.cursor()

            cursorObject.execute("""
                    CREATE TABLE IF NOT EXISTS employees (
                        id INT AUTO_INCREMENT PRIMARY KEY,
                        name VARCHAR(50),
                        age INT,
                        position VARCHAR(20),
                        salary FLOAT
                    )
                """)
            # print("Employee table created successfully!")
            cursorObject.close()
        except mysql.connector.Error as err:
            print(f"Error: {err}")
            return False

    def get_employee_details_from_user(self):
        try:
            name = input("Enter employee name: ")
            age = int(input("Enter employee age: "))
            position = input("Enter employee position: ")
            salary = float(input("Enter employee salary: "))
            return Employee(name, age, None, position, salary)
        except ValueError:
            print("Invalid input. Please enter valid details.")

    def add_employee(self, employee):
        try:
            cursor = self.connection.cursor()
            insert_query = "INSERT INTO employees (name, age, position, salary) VALUES (%s, %s, %s, %s)"
            values = (employee.name, employee.age, employee.position, employee.salary)
            cursor.execute(insert_query, values)
            self.connection.commit()
            cursor.close()
            print("Employee added successfully!")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def remove_employee_by_id(self, id):
        try:
            cursor = self.connection.cursor()

            # Check if employee with the ID exists
            cursor.execute("SELECT * FROM employees WHERE id = %s", (id,))
            employee = cursor.fetchone()

            if not employee:
                print("ID does not exists.")
                return

            delete_query = "DELETE FROM employees WHERE id = %s"
            cursor.execute(delete_query, (id,))
            self.connection.commit()
            cursor.close()
            print("Employee removed successfully!")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def remove_employee_by_name(self, name):
        try:
            cursor = self.connection.cursor()

            # Check if employee with the name exists
            cursor.execute("SELECT * FROM employees WHERE name = %s", (name,))
            employee = cursor.fetchone()

            if not employee:
                print("Name does not exists.")
                return

            delete_query = "DELETE FROM employees WHERE name = %s"
            cursor.execute(delete_query, (name,))
            self.connection.commit()
            cursor.close()
            print("Employee removed successfully!")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def update_employee_by_id(self, id, position, salary):
        try:
            cursor = self.connection.cursor()

            # Check if employee with the id exists
            cursor.execute("SELECT * FROM employees WHERE id = %s", (id,))
            employee = cursor.fetchone()

            if not employee:
                print("ID does not exists.")
                return

            update_query = "UPDATE employees SET position = %s, salary = %s WHERE id = %s"
            values = (position, salary, id)
            cursor.execute(update_query, values)
            self.connection.commit()
            cursor.close()
            print("Employee updated successfully!")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def update_employee_by_name(self, name, position, salary):
        try:
            cursor = self.connection.cursor()

            # Check if employee with the name exists
            cursor.execute("SELECT * FROM employees WHERE name = %s", (name,))
            employee = cursor.fetchone()

            if not employee:
                print("Name does not exists.")
                return

            update_query = "UPDATE employees SET position = %s, salary = %s WHERE name = %s"
            values = (position, salary, name)
            cursor.execute(update_query, values)
            self.connection.commit()
            cursor.close()
            print("Employee updated successfully!")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def display_all(self):
        try:
            cursor = self.connection.cursor()
            cursor.execute("SELECT * FROM employees")
            employees = cursor.fetchall()
            cursor.close()
            if employees:
                for emp in employees:
                    employee = Employee(emp[1], emp[2], emp[0], emp[3], emp[4])
                    employee.display_info()
            else:
                print("No employees found.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")

    def display_by_id(self, id):
        try:
            cursor = self.connection.cursor()
            cursor.execute("SELECT * FROM employees WHERE id = %s", (id,))
            employee = cursor.fetchone()
            cursor.close()
            if employee:
                emp = Employee(employee[1], employee[2], employee[0], employee[3], employee[4])
                emp.display_info()
            else:
                print("Employee not found.")
        except mysql.connector.Error as err:
            print(f"Error: {err}")


emp_manager = EmployeeManagement()

if emp_manager.connection.is_connected():
    emp_manager.display_all()


while True:
  print("----------------------------------")
  print("Welcome to the Employee Management Case Study")
  print("\nMenu:")
  print("1. Add Employee")
  print("2. Remove Employee by ID")
  print("3. Remove Employee by Name")
  print("4. Update Employee by ID")
  print("5. Update Employee by Name")
  print("6. Display all Employees")
  print("7. Display employee by ID")
  print("8. Exit")
  print("**************************************")

  user_input = int(input("Please set an option from 1 to 8: "))
  if user_input == 1:
    # Adding new employee
    employee = emp_manager.get_employee_details_from_user()
    emp_manager.add_employee(employee)

  elif user_input == 2:
    # Remove Employee record by ID
    id = int(input("Please enter the ID of the employee to remove: "))
    emp_manager.remove_employee_by_id(id)
  elif user_input == 3:
    # Remove Employee record by name
    name = input("Please enter the name of the employee to remove: ")
    emp_manager.remove_employee_by_name(name)

  elif user_input == 4:
    # Update Employee details by ID
    id = int(input("Please enter the ID of the employee to update: "))
    position = input("Please enter the new position: ")
    salary = float(input("Please enter the new salary: "))
    emp_manager.update_employee_by_id(id, position, salary)

  elif user_input == 5:
    # Update Employee details by name
    name = input("Please enter the name of the employee to update: ")
    position = input("Please enter the new position: ")
    salary = float(input("Please enter the new salary: "))
    emp_manager.update_employee_by_name(name, position, salary)

  elif user_input == 6:
    # Display all Employees
    emp_manager.display_all()

  elif user_input == 7:
    # Display employee by ID
    id = int(input("Please enter the ID of the employee to display: "))
    emp_manager.display_by_id(id)

  elif user_input == 8:
    print("Exiting the Employee Management Case Study.")
    break

  else:
    print("Invalid option. Please select an option between (1-8).")
```
